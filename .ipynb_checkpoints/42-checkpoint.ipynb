```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

df_wine = datasets.load_wine()

model_KMeans = KMeans(n_clusters=3).fit(df_wine.data) 
predictions = model_KMeans.predict(df_wine.data)

pca = PCA(n_components=2).fit(df_wine.data)
wine_pca = pca.fit_transform(df_wine.data)

sns.set_style('white')
plt.figure()
sns.scatterplot(hue=predictions, x=wine_pca[:,0], y=wine_pca[:,1], palette="bright")
plt.title("Кластеризація методом k-середніх", fontsize=15)
plt.show()
```

```python
model_dbscan = DBSCAN(eps=50, min_samples=8)
model_dbscan_predictions = model_dbscan.fit(df_wine.data)

pca = PCA(n_components=2).fit(df_wine.data)
wine_pca = pca.fit_transform(df_wine.data)

sns.set_style('white')
plt.figure()
sns.scatterplot(hue=model_dbscan_predictions.labels_, x=wine_pca[:,0], y=wine_pca[:,1], palette="bright")
plt.title("Кластеризація на основі щільності (DBSCAN)", fontsize=15)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

mergings = linkage(df_wine.data, method='average', optimal_ordering=True)

dendrogram(mergings, truncate_mode="level", p=1)

plt.title('Ієрархічна кластеризація', fontsize=15)
plt.xlabel('Кількість точок у вузлі', fontsize=15)
plt.show()
```
